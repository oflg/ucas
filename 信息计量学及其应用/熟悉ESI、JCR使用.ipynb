# 熟悉 ESI、JCR 使用


## 清华大学、北京大学、中国科学院大学、浙江大学进入ESI的学科数量。

### 答案
清华大学 21 个、北京大学 22 个、中国科学院大学 21 个、浙江大学 21 个。

### 过程
在 ESI 侧边栏的“Filter Results By”中点击“Add Filter”按钮，再点击“Institutions”按钮，选择 TSINGHUA UNIVERSITY，可得清华大学的“Research Fields”中学科数量为 21 个。

![](熟悉ESI、JCR使用/img/2022-09-25-21-19-27.png)

同理可得，北京大学 22 个，中国科学院大学 21 个，浙江大学 21 个。


## 进入ESI的期刊，按照篇均被引、篇均top paper排在前5位的期刊都有哪些？是否有重合？有什么发现？

### 答案
篇均被引前 5： CA-A CANCER JOURNAL FOR CLINICIANS、ANNUAL REVIEW OF ASTRONOMY AND ASTROPHYSICS, VOL 51、ANNUAL REVIEW OF PSYCHOLOGY, VOL 63、ANNUAL REVIEW OF IMMUNOLOGY, VOL 31 和 ANNUAL REVIEW OF PSYCHOLOGY, VOL 64。

篇均 top paper 前 5：NATURE、NATURE COMMUNICATIONS、SCIENCE、JOURNAL OF THE AMERICAN CHEMICAL SOCIETY 和 ANGEWANDTE CHEMIE-INTERNATIONAL EDITION。

没有重合。

发现不能只把篇均被引量作为评价顶刊的标准，顶刊往往篇均被引量不高。

### 过程
在 ESI 侧边栏的“Results List”中选择“Journals”，再在表头中点击“Cites/Paper”按钮。

![](熟悉ESI、JCR使用/img/2022-09-25-11-03-32.png)

得到篇均被引前 5 位期刊为： CA-A CANCER JOURNAL FOR CLINICIANS、ANNUAL REVIEW OF ASTRONOMY AND ASTROPHYSICS, VOL 51、ANNUAL REVIEW OF PSYCHOLOGY, VOL 63、ANNUAL REVIEW OF IMMUNOLOGY, VOL 31 和 ANNUAL REVIEW OF PSYCHOLOGY, VOL 64。

同理，在表头中点击“Top Papers”按钮，得到篇均 top paper 前 5 位期刊为：NATURE、NATURE COMMUNICATIONS、SCIENCE、JOURNAL OF THE AMERICAN CHEMICAL SOCIETY 和 ANGEWANDTE CHEMIE-INTERNATIONAL EDITION。



## 请对比材料学领域 citation rates与前50%被引次数的异同。

### 答案
citation rates 为 18.92，前 50% 被引次数为 8，体现了马太效应。

### 过程
在 ESI 顶栏中点击“Field Baselines”，在表格中查看“MATERIALS SCIENCE”行“All Years”列数据为 18.92，即为材料学领域 citation rates。

![](熟悉ESI、JCR使用/img/2022-09-25-11-18-40.png)

在侧边栏中点击“Percentiles”，在表格中找到“MATERIALS SCIENCE”类，查看中“50.00%”行“All Years”列数据为 8，即为前 50% 被引次数。

![](熟悉ESI、JCR使用/img/2022-09-25-11-19-45.png)



## 材料学领域Highly Cited Thresholds与前1%被引次数的异同。

### 答案
Highly Cited Thresholds 与前 1% 被引次数在数值上相同。高被引论文（Highly Cited Paper）是同一年同一个 ESI 学科中发表的所有论文按被引用次数由高到低进行排序，排在前1% 的论文。

### 过程

在 ESI 顶栏中点击“Field Baselines”，在侧边栏中点击“Percentiles”，在表格中找到“MATERIALS SCIENCE”类，查看“1.00%”行数据如下：

![](熟悉ESI、JCR使用/img/2022-09-25-14-15-51.png)

在 ESI 顶栏中点击“Citation Thresholds”，在侧边栏中点击“Highly Cited Thresholds”，在表格中找到“MATERIALS SCIENCE”类，查看“10.00%”行数据如下：

![](熟悉ESI、JCR使用/img/2022-09-25-14-07-56.png)

两行数据完全相同。



## 中国大陆、美国、英国、日本2016-2021年进入SCIE期刊的数量，在4个分区的数量？各国影响因子最高的期刊都是谁？

### 答案

| 国家 | SCIE期刊的数量 | Q1 区 | Q2 区 | Q3 区 | Q4 区 | 影响因子最高期刊 |
|---|---|---|---|---|---|---|
| 中国大陆 | 273 | 123 | 84 | 58 | 37 | CELL RESEARCH |
| 美国 | 3112 | 988 | 1029 | 1008 | 777 | CA-A CANCER JOURNAL FOR CLINICIANS |
| 英国 | 2087 | 754 | 737 | 657 | 395 | LANCET |
| 日本 | 245 | 15 | 46 | 93 | 129 | JOURNAL OF PHOTOCHEMISTRY AND PHOTOBIOLOGY C-PHOTOCHEMISTRY REVIEWS |


### 过程

在 JCR 顶栏中点击“Journals”，在左侧点击“Filter”悬浮按钮，点击“Citation Indexes”选择“SCIE”，点击“Country/region"选择“CHINA MAINLAND”再点击“APPLY”按钮进行筛选，可得中国大陆期刊数量为 273。在数据表头处点击“2021 JIF”按期刊影响因子排序，可得中国大陆影响因子最高的期刊为“CELL RESEARCH”。

点击“Filter”悬浮按钮，在“JIF Quartile"中分别选择“Q1”、“Q2”、“Q3”、“Q4”，可得 Q1、Q2、Q3、Q4 区期刊数量分别为 123、84、58、37。

![](熟悉ESI、JCR使用/img/2022-09-25-14-55-24.png)

同理，分别更改“Country/region"为“USA”、“NORTHERN IRELAND、ENGLAN、SCOTLAND、WALES”和“JAPAN”，可得美国、英国和日本数据。




## 全部期刊中2021年影响因子最高的5本期刊是谁？按照ESI学科分类，材料学领域影响因子最高的期刊是？出版社是？它的2021，2020，2019年普赖斯指数是？它的自引率？

### 答案

2021年影响因子最高的5本期刊：CA-A CANCER JOURNAL FOR CLINICIANS、LANCET、NEW ENGLAND JOURNAL OF MEDICINE、JAMA-JOURNAL OF THE AMERICAN MEDICAL ASSOCIATION、NATURE REVIEWS MOLECULAR CELL BIOLOGY。

材料学领域影响因子最高的期刊是 Applied Sciences-Basel，其出版社是 MDPI，其 2021、2020、2019年普赖斯指数分别是 2.838、2.679、2.474，其自引率为 12.95% 。

### 过程

在 JCR 顶栏中点击“Journals”，在数据表头处点击“2021 JIF”按期刊影响因子排序，可得全部期刊中2021年影响因子最高的5本期刊为：CA-A CANCER JOURNAL FOR CLINICIANS、LANCET、NEW ENGLAND JOURNAL OF MEDICINE、JAMA-JOURNAL OF THE AMERICAN MEDICAL ASSOCIATION、NATURE REVIEWS MOLECULAR CELL BIOLOGY。

![](熟悉ESI、JCR使用/img/2022-09-25-15-36-43.png)

在 JCR 顶栏中点击“Journals”，在左侧点击“Filter”悬浮按钮，点击“Categories”勾选带“Materials Science”的学科，再点击“APPLY”按钮进行筛选。可得到材料学领域影响因子最高的期刊是 Applied Sciences-Basel。

![](熟悉ESI、JCR使用/img/2022-09-25-15-49-17.png)

点击“Applied Sciences-Basel”查看期刊详情。

在“Journal information”处，可以看到其出版社是 MDPI：

![](熟悉ESI、JCR使用/img/2022-09-25-15-51-54.png)

在“Journal Impact Factor”处，可以看到2021、2020、2019年普赖斯指数（期刊影响因子）分别是 2.838、2.679、2.474：

![](熟悉ESI、JCR使用/img/2022-09-25-16-15-13.png)

在“Cited Half-life”处，可以看到期刊被引用的总次数（TOTAL NUMBER OF CITES）为 63761，被本刊引用的次数（SELF-CITATIONS）为 8258。自引率 ＝被本刊引用的次数/期刊被引用的总次数 ＝ 12.95% 。

![](熟悉ESI、JCR使用/img/2022-09-25-16-24-29.png)



## 全部期刊中规范化的特征因子最高的5本期刊是谁？按照ESI学科分类，材料学领域规范化的特征因子最高的期刊是？与影响因子高的期刊对比？

### 答案

全部期刊中规范化的特征因子最高的5本期刊是：Nature Communications、Scientific Reports、NATURE、SCIENCE、PLoS One。

材料学领域规范化的特征因子最高的期刊是：ADVANCED MATERIALS。

影响因子最高的 5 本期刊对比发现，规范化的特征因子最高的 5 本期刊与其**没有重合**。

### 过程

在 JCR 顶栏中点击“Journals”，在数据表上方点击“Customize”在弹窗中仅勾选“Eigenfactor”，再点击“APPLY”按钮进行筛选，可得全部期刊中规范化的特征因子最高的5本期刊为：Nature Communications、Scientific Reports、NATURE、SCIENCE、PLoS One。

![](熟悉ESI、JCR使用/img/2022-09-25-16-59-25.png)

在左侧点击“Filter”悬浮按钮，点击“Categories”勾选带“Materials Science”的学科，再点击“APPLY”按钮进行筛选，可得材料学领域规范化的特征因子最高的期刊是：ADVANCED MATERIALS。

![](熟悉ESI、JCR使用/img/2022-09-25-17-03-12.png)

全部期刊中2021年JCI最高的5本期刊是：CA-A CANCER JOURNAL FOR CLINICIANS、NEW ENGLAND JOURNAL OF MEDICINE、LANCET、MMWR Surveillance Summaries、Lancet Respiratory Medicine。

物理学领域 JCI 最高的期刊是：Nature Photonics，其近5年影响因子是 40.182，其被引和施引半衰期分别是 7.0、5.9。


### 过程

![](熟悉ESI、JCR使用/img/2022-09-25-17-10-09.png)

在左侧点击“Filter”悬浮按钮，点击“Categories”勾选带“Physics”的学科，再点击“APPLY”按钮进行筛选，可得物理学领域 JCI 最高的期刊是：Nature Photonics。

![](熟悉ESI、JCR使用/img/2022-09-25-21-07-48.png)


点击“Nature Photonics”查看期刊详情。

在“5 Year Impact Factor”处，可以看到其近5年影响因子是 40.182：

![](熟悉ESI、JCR使用/img/2022-09-25-21-12-25.png)
